```python
cdf2 = cdf.drop(columns=['track_id', 'artists', 'album_name','track_name', 'explicit', 'mode', 'track_genre'])
```

```python
corr_matrix = cdf2.corr()
```

```python
px.imshow(
    corr_matrix,
    text_auto=True,
    aspect="auto",
    color_continuous_scale="RdBu_r",
    title="Correlation Matrix Heatmap")
```

```python
df4 = cdf50.drop(columns=['duration_ms', 'key', 'speechiness', 'liveness', 'tempo', 'time_signature'])
df4.head()
```
